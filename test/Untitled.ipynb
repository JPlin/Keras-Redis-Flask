```python
# cmd line
# python simple_request.py

# import the necessary packages
import requests
import numpy as np
import base64
import matplotlib.pyplot as plt

# initialize the Keras REST API endpoint URL along with the input
# image path
KERAS_REST_API_URL = "http://localhost:9191/parsing"
#KERAS_REST_API_URL = "https://paringweb.herokuapp.com/classify"
IMAGE_PATH = "../static/img/13601661_1_00_img.png"

# load the input image and construct the payload for the request
image = open(IMAGE_PATH, "rb").read()
payload = {"image": image}

# submit the request
r = requests.post(KERAS_REST_API_URL, files=payload).json()
```

```python
image = r['predictions']['image']
```

```python
# ensure the request was sucessful
a = np.frombuffer(base64.decodestring(image.encode('utf-8')), dtype=np.uint8)
a = a.reshape((512,512,3))
plt.imshow(a.astype(np.uint8))
plt.show()
```
